# Finding peak using peakutils module Tutorial !!

###### We will us a dataset which have 10 peaks and we will find the index of all the 10 peaks !!!!!!!!

```python
# Necessary imports !!

import pandas as pd
import numpy as np
import peakutils
from peakutils.plot import plot as pplot
from matplotlib import pyplot
from numpy import linalg as LA
```

###### Importing a local file !!

```python
dataframe = pd.read_csv('10_step_count.csv')
dataset = dataframe.values

# dataset 6th 7th and 8th index are x,y and z axis !!!
# so we will find the magnitude i.e. norm of the vector (x,y,z) !!

accn_mag = LA.norm(dataset[:,6:9], axis = 1)
```

```python
# We will separate equal interval from 0 to length of accn_mag with number of division as lenght of accn_mag !!!
x = np.linspace(0, len(accn_mag), len(accn_mag))
y = accn_mag
pyplot.figure(figsize=(10,6))
pyplot.plot(x, y)
pyplot.title(" plot ")
```

##### Now we see their are 10 peaks !!!!
##### So peakutils module will get all the 10 peaks for us !!

```python
indexes = peakutils.indexes(y, thres=0.5, min_dist=15)
print(indexes)
print(x[indexes], y[indexes])
pyplot.figure(figsize=(10,6))
pplot(x, y, indexes)
pyplot.title('First estimate')
```

###### All the 10 peaks are shown with red points !!!
